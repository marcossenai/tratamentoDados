```python
#Exibir as 5 primeiras linhas
import pandas as pd

df = pd.read_excel("dados_ausentes.xlsx")

display(df.head(5))
```

```python
#Verificar se há valores nulos
display(df.isnull())
```

```python
#Substituir dados ausentes

df['Salário'].fillna(0, inplace=True)
display(df)
```

```python
#Substituir os valores NaN de uma coluna por um valor específico

df['Idade'].fillna(df['Idade'].mean(), inplace=True)
display(df)
```

```python
#Excluir as linhas com dados ausentes
df_sem_nulos = df.dropna()
display(df_sem_nulos)
```

```python
df_cidade_sem_nulos = df.dropna(subset=['Cidade'])
print(df_cidade_sem_nulos)
```
