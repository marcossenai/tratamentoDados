```python
# 1 Exibir as 5 primeiras linhas
import pandas as pd

df = pd.read_excel("dados_ausentes.xlsx")

display(df.head(5))
```

```python
# 2 Verificar se há valores nulos
display(df.isnull())
```

```python
# 3 Substituir dados ausentes

df['Salário'].fillna(0, inplace=True)
display(df)
```

```python
# 4 Substituir os valores NaN de uma coluna por um valor específico

df['Idade'].fillna(df['Idade'].mean(), inplace=True)
display(df)
```

```python
# 5 Excluir as linhas com dados ausentes
df_sem_nulos = df.dropna()
display(df_sem_nulos)
```

```python
#6

df_cidade_sem_nulos = df.dropna(subset=['Cidade'])
print(df_cidade_sem_nulos)
```

```python
#7

df_2 = pd.read_excel("dados_ausentes.xlsx")

df_2['Idade'].fillna(df_2['Idade'].median(), inplace=True)

display(df_2)
```

```python
#8 Substituir valores NaN de uma coluna com o valor de uma outra coluna

df_3 = pd.read_excel("dados_ausentes.xlsx")

df_3['Salário'].fillna(df_3['Idade'], inplace=True)

display(df_3)
```

```python
# 9  Exercício: Preencher valores NaN com o valor anterior (método de forward fill)

df_2['Salário'].fillna(df_2['Salário'].ffill(), inplace=True)

display(df_2)
```

```python
 #10 Exercício: Preencher valores NaN com o valor posterior (método de backward fill)

df_2['Cidade'].fillna(df_2['Cidade'].bfill(), inplace=True)

display(df_2)
```

```python
#11 Exercício: Alterar todos os valores de uma coluna para maiúsculas.

df_2['Cidade'] = df_2['Cidade'].str.title()

display(df_2)
```

```python
#12 Exercício: Alterar todos os valores de uma coluna para minúsculas.

df_2['Cidade'] = df_2['Cidade'].str.lower()

display(df_2)
```

```python
# 13. Exercício: Remover espaços em branco à esquerda e à direita em uma coluna

df_3['Cidade'] = df_3['Cidade'].str.strip()
display(df_3)
```

```python
# 14 Exercício: Substituir valores de uma coluna com base em um dicionário de mapeamento (ex: 'São Paulo' para 'SP').

df_2['Cidade'] = df_2['Cidade'].replace({'são paulo': "SP", 'rio de janeiro': 'RJ', 'rj': 'RJ'})

display(df_2)
```

```python
# 15 Identificar e remover valores duplicados de um DataFrame.

df_3= df_3.drop_duplicates()
display(df_3)
```
